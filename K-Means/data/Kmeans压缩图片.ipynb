```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
bird = plt.imread('../data/bird_small.png')
bird.shape
```

```python
plt.imshow(bird)
```

```python
kmean = KMeans(4)
```

```python
kmean.fit(bird)
```

```python
X = bird.reshape(-1,3)
```

```python
X.shape
```

```python
X
```

```python
kmean.fit(X)

```

```python
y_ = kmean.predict(X)
y_
```

```python
# 
colors  = kmean.cluster_centers_
```

```python
plt.imshow(bird)
```

```python
image = colors[y_]

image.shape

plt.imshow(image.reshape(128,128,3))
```

```python
kmean = KMeans(16)

kmean.fit(X)

y_ = kmean.predict(X)
y_
```

```python
colors = kmean.cluster_centers_
colors
```

```python
image = colors[y_]

image.shape

plt.imshow(image.reshape(128,128,3))
```
